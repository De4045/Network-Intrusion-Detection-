# Step 2: Define a function to calculate outlier percentages for a given column
def outlier_percentage(df, num_cols):
    """
    Calculates the percentage of outliers for each numeric column in a DataFrame
    using the Interquartile Range (IQR) method.

    Args:
        df (pd.DataFrame): The input DataFrame.
        num_cols (pd.Index): A list of numeric columns to check for outliers.

    Returns:
        pd.Series: A Series containing the percentage of outliers for each column.
    """
    outlier_info = {}
    for col in num_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Boolean mask to identify outliers
        outliers = (df[col] < lower_bound) | (df[col] > upper_bound)
        
        # Calculate the percentage
        percent = (outliers.sum() / len(df[col])) * 100
        outlier_info[col] = round(percent, 2)
    return pd.Series(outlier_info, name='Outlier_Percentage')

# Step 3: Run the function and print the results
result = outlier_percentage(df, num_cols)
print(result)

# Optional: Print the results in a formatted markdown table
print("\n" + "=" * 30)
print("Outlier Percentage Table:")
print("=" * 30)
outlier_df = result.to_frame().reset_index()
outlier_df.columns = ['Column Name', 'Outlier Percentage']
print(outlier_df.to_markdown(index=False))

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec
# Algorithms
algorithms = [
    'MULTI-CLASS CLASSIFICATION',
    'DECISION TREE',
    'RANDOM FOREST',
    'NAIVE BAYES',
    'SUPPORT VECTOR CLASSIFIER',
    'KNN'
]
# Label Data
Actual1 = ["Normal"] * 6
Prediction1 = ["Normal"] * 6
Actual2 = ["R2L"] * 6
Prediction2 = ["R2L"] * 6
Actual3 = ["DoS"] * 6
Prediction3 = ["Normal", "DoS", "DoS", "DoS", "DoS", "DoS"] # One misclassified
# Accuracy values
accuracies = [score1, score2, score3, score4, score5, score6]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec

# Mapping Labels
label_map = {"Normal": 0, "R2L": 1, "DoS": 2}

# Prepare DataFrame (assuming 'algorithms', 'Actual1', 'Prediction1', etc. are defined elsewhere)
data = []
for i, alg in enumerate(algorithms):
    data.extend([
        {'Algorithm': alg, 'Type': 'Actual1', 'Label': Actual1[i], 'value': label_map[Actual1[i]]},
        {'Algorithm': alg, 'Type': 'Predicted1', 'Label': Prediction1[i], 'value': label_map[Prediction1[i]]},
        {'Algorithm': alg, 'Type': 'Actual2', 'Label': Actual2[i], 'value': label_map[Actual2[i]]},
        {'Algorithm': alg, 'Type': 'Predicted2', 'Label': Prediction2[i], 'value': label_map[Prediction2[i]]},
        {'Algorithm': alg, 'Type': 'Actual3', 'Label': Actual3[i], 'value': label_map[Actual3[i]]},
        {'Algorithm': alg, 'Type': 'Predicted3', 'Label': Prediction3[i], 'value': label_map[Prediction3[i]]}
    ])
df = pd.DataFrame(data)
df['Plotvalue'] = df['value'] + 0.1

# Set up two vertical subplots
fig = plt.figure(figsize=(16, 6))
gs = GridSpec(1, 2, width_ratios=[3, 1])

# Left Plot: Actual vs Predicted Labels
ax1 = fig.add_subplot(gs[0])
sns.barplot(data=df, x='Algorithm', y='Plotvalue', hue='Type', palette='Set2', ax=ax1)

# Annotate Labels
for p in ax1.patches:
    height = p.get_height()
    orig_val = int(round(height - 0.1))
    label_text = list(label_map.keys())[list(label_map.values()).index(orig_val)]
    ax1.text(p.get_x() + p.get_width() / 2, height + 0.05, label_text, ha='center', fontsize=8)

ax1.set_ylabel("Values")
ax1.set_title('Actual vs Predicted Labels')
ax1.set_ylim(0, 3)
ax1.tick_params(axis='x', rotation=15)
ax1.legend(loc='upper left', bbox_to_anchor=(1, 1))

# ---- Right Plot: Accuracy (Vertical) ----
ax2 = fig.add_subplot(gs[1])
sns.barplot(x=algorithms, y=accuracies, palette='coolwarm', ax=ax2)

# Add accuracy Labels on top
for i, acc in enumerate(accuracies):
    ax2.text(i, acc + 0.01, f"{acc:.3f}", ha='center', fontsize=9)

ax2.set_ylim(0, 1.05)
ax2.set_title('Accuracy per Algorithm')
ax2.set_ylabel('Accuracy')
ax2.tick_params(axis='x', rotation=90)

# Final Layout
plt.tight_layout()
plt.show()
```
